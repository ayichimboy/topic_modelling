```python
def find_prime(nums):
    prime = []
    
    for n in nums:
        if n<2:
            continue
        is_prime = True
        
        for i in range(2, int(n**0.5) + 1):
            if n%i == 0:
                is_prime = False
                break
            if is_prime:
                prime.append(n)
        return prime
```

```python
def mysqrt(x):
    
    if x < 2:
        return x
    
    left, right = 1, x//2
    
    while left <= right:
        mid = (left + right) //2
        if mid * mid == x:
            return mid
        elif mid*mid < x:
            left = mid + 1
        else:
            right = mid -1
            
    return right

mysqrt(8)
```

```python
# find sqrt of number 

def sqrme(x):
    
    if x < 2:
        return x
    
    left_val = 1
    right_val = x//2
    
    while left_val <= right_val:
        mid_val = (left_val+right_val)//2
        
        if mid_val*mid_val == x:
            return mid_val
        elif mid_val*mid_val < x:
            left_val = mid_val +1
        else:
            right_val = mid_val-1
            
    return right_val
        
        
sqrme(8)   
        
        
```

```python
# majority number 
mynum = [3,2,3]

def majorNum(nums):
    createDict = {}

    for n in nums:
         createDict[n] = createDict.get(n,0) + 1
     
    # mydict = [val/2 for val, key in createDict.items()]      
    max_key = max(createDict.values())
    for key, val in createDict.items():
        if val == max_key:
            return key

majorNum(mynum)     
          
```
